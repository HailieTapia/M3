```python
import pandas as pd

ruta = r'C:\Users\htapi\Downloads\copia\doc\transaccionesFilograficos.csv'

with open(ruta, 'r', encoding='utf-8') as f:
    transactions = [line.strip().split(',') for line in f if line.strip()]
max_items = max(len(t) for t in transactions)
transactions_padded = [t + [None]*(max_items - len(t)) for t in transactions]
df = pd.DataFrame(transactions_padded, columns=list(range(max_items)))
print(df.shape)
```

```python
transactions = []
for i in range(df.shape[0]):
    # Convertimos a string y filtramos los None (que se ven como 'None' o 'nan' como string)
    transaction = [str(item) for item in df.iloc[i] if pd.notna(item)]
    transactions.append(transaction)

transactions # Mostrar las primeras 5 transacciones para validar
```

```python
from mlxtend.preprocessing import TransactionEncoder
# Crear un DataFrame con todas las transacciones en formato booleano
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_ary, columns=te.columns_)

print(df.head())
```

```python
from mlxtend.frequent_patterns import apriori

# Generación de conjuntos de ítems frecuentes con soporte mínimo 1%
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)

# Mostrar los primeros ítems frecuentes
print(frequent_itemsets.head())
```

```python
from mlxtend.frequent_patterns import association_rules

# Generar reglas de asociación usando 'lift' con umbral mínimo 1
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Mostrar las primeras reglas generadas
print(rules.head())
```

```python
# Ordenar reglas por confianza descendente
rules = rules.sort_values(by='confidence', ascending=False)

# Mostrar las 20 reglas con mayor confianza
print(rules.head(20))
```

```python
import joblib

# Guardar las reglas en un archivo pkl
joblib.dump(rules, 'reglas_asociacion.pkl')
```
